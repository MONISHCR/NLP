**Word Tokenize using spaCy**

```python
import spacy

# Load English language model
nlp = spacy.load("en_core_web_sm")

def process_text(text):
    # Process the text using spaCy
    doc = nlp(text)

    # Word tokens
    word_tokens = [token.text for token in doc]
    print("Word tokens:")
    print(word_tokens)

    # Base Words (lemmatization)
    lemmatized = [token.lemma_ for token in doc]
    print("Base Words:")
    print(lemmatized)

    # Parts of Speech
    pos = [token.pos_ for token in doc]
    print("Parts of Speech:")
    print(pos)

text = "Keshav Memorial Institute of Technology (KMIT) is a prestigious educational institution located in Hyderabad, India. Renowned for its commitment to academic excellence and innovation, KMIT has established itself as a leading technical institute in the region. Equipped with state-of-the-art facilities and a dedicated faculty, the institution provides a conducive learning environment for students to thrive. With a focus on practical learning and industry exposure, KMIT prepares its students to meet the demands of the ever-evolving technological landscape. The institute encourages research and innovation, fostering a culture of creativity and critical thinking among its students. Beyond academics, KMIT promotes holistic development through a wide range of extracurricular activities and initiatives. With a strong emphasis on ethics and values, the institution aims to mold individuals who not only excel in their chosen fields but also contribute positively to society. Through partnerships with leading industries and collaborations with academic institutions worldwide, KMIT ensures that its graduates are well-equipped to succeed in the global marketplace. In essence, Keshav Memorial Institute of Technology is not just a center for education; it's a hub of innovation and excellence, shaping the future leaders and pioneers of tomorrow."

# Call the function to process the text
process_text(text)
```

**Sentence Tokenization using spaCy**



```python
doc = nlp(text)

# Extract sentence tokens
sentence_tokens = [sent.text for sent in doc.sents]

# Print the sentence tokens
print("Sentence tokens:")
for token in sentence_tokens:
    print(token)
```



**Removing Stop Words**

```python
# Removing Stop Words
print("Removing Stop Words:")
print("...................................................")
text_without_stopwords = ' '.join([token.text for token in doc if not token.is_stop])
print("The stop words in the paragraph are:")
stop_words_in_para = [token for token in doc if token.is_stop]
print(stop_words_in_para)
print("...................................................")
print("Total number of stop words:")
print(len(stop_words_in_para))
print("...................................................")
print("Text With Stop Words:")
print(text)
print("...................................................")
print("Text Without Stop Words:")
print(text_without_stopwords)
```

**POS tagging**

```python
# POS Tagging after removing stop words
print("POS Tagging after removing stop words:")
print("...................................................")
pos_tags = [(token.text, token.pos_) for token in doc if not token.is_stop]
print("Word\t\tPOS Tag")
print("---------------------------")
for word, pos_tag in pos_tags:
    print(f"{word}\t\t{pos_tag}")
print("...................................................")

print("Text With Stop Words:")
print(text)
print("...................................................")
print("Text Without Stop Words:")
print(text_without_stopwords)
```

**Dependency Parsing**

```python
from spacy import displacy
displacy.render(doc, style="dep", options={'distance': 70})
```
